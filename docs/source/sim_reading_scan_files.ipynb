# Reading Simulated Scan Data Files

LISO tries to keep the API of reading simulated data akin to that of reading experimental data as much as possible. In experiments, the *macropulse ID* is the only identity that we can trust to align data from different channels. However, we do not have the same issue in simulations. Nevertheless, we still introduced the *simulation ID* which is the counter part of the *macropulse ID* and starts from 1 by default.

```python
from liso import open_sim
```

```python
sim = open_sim("../../examples/astra_scan/")

sim.info()
```

## Obtain general information

```python
sim.sim_ids
```

```python
sim.control_channels
```

```python
sim.phasespace_channels
```

## Access a single simulation result

```python
# From index
sid, data = sim.from_index(2)
assert sid == sim.sim_ids[2]

data
```

```python
# From simulation ID
sid, data = sim.from_id(1)
assert sid == 1
```

## Access channel data

#### Control channel

```python
gun_gradient_channel = sim.channel("gun/gun_gradient")
```

```python
# convert the channel data to numpy array
gun_gradient_channel.numpy()
```

#### Phasespace channel

```python
out_x_channel = sim.channel("gun/out", 'x')
```

```python
# convert the channel data to numpy array
out_x_channel.numpy()
```
